```python
def show_countplot(data, col, title, figsize=(8, 6)):
    sns.set(style="dark", rc={
        'figure.figsize':figsize,
        'axes.labelsize':16,
        'xtick.labelsize':12,
        'ytick.labelsize':12,
        'xtick.color':"black",
        'ytick.color':"black"
    })
    
    ax = sns.countplot(
        x=col,
        data=data,
        edgecolor=(1, 1, 1),
        palette=sns.color_palette('Paired', 2),
        linewidth=2
    )
    ax.set_title(title, fontsize=18)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=60, ha="right")
    plt.show()
```

```python
show_countplot(hr, "dept", "Existing Employees")
show_countplot(hr1, "dept", "Left Employees")
```

```python
def show_satisfaction_level(data, title):
    sns.set(style='dark', rc={
        'figure.figsize':(8, 6),
        'axes.labelsize':16,
        'xtick.labelsize':12,
        'ytick.labelsize':12,
        'xtick.color':"black",
        'ytick.color':"black",
    })
    
    data = data["satisfaction_level"] \
        .value_counts() \
        .sort_index()
    
    ax = sns.lineplot(
        x=data.index,
        y=data.values,
        legend="full"
    )
    ax.set_title(title, fontsize=18)
    ax.set_xlabel("satisfaction_level")
    ax.set_ylabel("No. of Employees")
    
    plt.xticks(np.around(np.linspace(min(data.index), max(data.index), 20), 2))
    plt.show()

show_satisfaction_level(hr, "Existing Employees")
show_satisfaction_level(hr1, "Left Employees")
```

```python
def show_average_montly_hours(data, title):
    sns.set(style='dark', rc={
        'figure.figsize':(8, 6),
        'axes.labelsize':16,
        'xtick.labelsize':12,
        'ytick.labelsize':12,
        'xtick.color':"black",
        'ytick.color':"black",
    })
    
    data = data["average_montly_hours"] \
        .value_counts() \
        .sort_index()
    
    ax = sns.lineplot(
        x=data.index,
        y=data.values,
        legend="full"
    )
    ax.set_title(title, fontsize=18)
    ax.set_xlabel("average_montly_hours")
    ax.set_ylabel("No. of Employees")
    
    plt.xticks(np.around(np.linspace(min(data.index), max(data.index), 20), 2))
    plt.show()

show_average_montly_hours(hr, "Existing Employees")
show_average_montly_hours(hr1, "Left Employees")
```
